## TensorFlow Serving

TensorFlow offers a [serving platform](https://github.com/tensorflow/serving) that enables people to deliver their machine learning models to the outside world in a production environment. Models are made accessible via either HTTP or gRPC endpoints. The author of a model will use the TF Serving docker container and move their saved model into the container. After doing this, they can expose a port which can be used to pass data into the model and retrieve predictions.

Below is a demonstration of how my object detection model could be used in a production environment. The docker setup was done following this [tutorial](https://neptune.ai/blog/how-to-serve-machine-learning-models-with-tensorflow-serving-and-docker), but there are countless other examples available.

```python
def make_prediction(instances):
    data = json.dumps({"signature_name": "serving_default", "instances": instances.tolist()})
    headers = {"content-type": "application/json"}
    json_response = requests.post(url, data=data, headers=headers)
    predictions = json.loads(json_response.text)#['predictions']
    return predictions
```

```python
def extract_bboxes(mask_path):
    mask = cv2.imread(mask_path)
    mask = cv2.resize(mask, (250, 250))
    mask[mask != 0] = 1

    m = mask[:, :, 0]

    # Bounding box.

    horizontal_indicies = np.where(np.any(m, axis=0))[0]

    vertical_indicies = np.where(np.any(m, axis=1))[0]

    if horizontal_indicies.shape[0]:

        x1, x2 = horizontal_indicies[[0, -1]]

        y1, y2 = vertical_indicies[[0, -1]]

        # x2 and y2 should not be part of the box. Increment by 1.

        x2 += 1

        y2 += 1

    else:

        # No mask for this instance. Might happen due to

        # resizing or cropping. Set bbox to zeros

        x1, x2, y1, y2 = 0, 0, 0, 0

    return [x1, y1, x2, y2]
```

```python
# Loop through images and extract predictions
preds = []
for img in img_paths:
    x_test = cv2.imread(img, cv2.IMREAD_COLOR)
    x_test = np.expand_dims(x_test, axis=0)
    preds.append(make_prediction(x_test)['predictions'][0])
```

```python
# Store only the prediction with the highest probability
outputs = {key: [] for key in preds[0].keys()}
keys = list(preds[0].keys())
for predictions in preds:
    for i in range(len(predictions['detection_boxes'])):
        if i > 0:
            break
        for key in predictions.keys():
            val = predictions[key][i] if isinstance(predictions[key], list) else predictions[key]
            outputs[key].append(val)
```

```python
# Plot examples of original image with the actual mask bounding box and the predicted box, along with classification
for i in range(len(preds)):
    im = cv2.imread(img_paths[i], cv2.IMREAD_COLOR)
    x1, y1, x2, y2 = extract_bboxes(mask_paths[i])
    width = x2 - x1
    height = y2 - y1
    im = np.array(im)
    mask_im = cv2.imread(mask_paths[i], cv2.IMREAD_COLOR)
    mask_im = cv2.resize(mask_im, (250, 250))

    # Create a Rectangle patch
    rect = patches.Rectangle((x1,y1),width,height,linewidth=1,edgecolor='r',facecolor='none')
    rect2 = patches.Rectangle((x1,y1),width,height,linewidth=1,edgecolor='r',facecolor='none')

    pred_x1, pred_y1, pred_x2, pred_y2 = [x*250 for x in outputs['raw_detection_boxes'][i]]
    pred_width, pred_height = pred_x2 - pred_x1, pred_y2 - pred_y1
    rect3 = patches.Rectangle((pred_x1,pred_y1),pred_width,pred_height,linewidth=1,edgecolor='blue',facecolor='none')

    # Create figure and axes
    fig,ax = plt.subplots(1,2, figsize=(12,12), dpi = 90)

    pred = label_map[int(outputs['detection_classes'][i])]
    actual = 'Malignant' if 'malignant' in img_paths[i].lower() else 'Benign'

    ax[0].title.set_text('Actual: {}\nPredicted {}'.format(actual, pred))
    ax[0].imshow(im)
    ax[0].add_patch(rect)
    ax[0].add_patch(rect3)
    ax[1].imshow(mask_im)
    ax[1].add_patch(rect2)
    plt.show()
```
